```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
COLORS = ["#348ABD", "#A60628", "#7A68A6", "#467821", "#D55E00", "#CC79A7", "#56B4E9", "#009E73", "#F0E442", "#0072B2"]
matplotlib.rcParams['axes.prop_cycle'] = matplotlib.cycler(color=COLORS) 
```

```python
import pandas as pd
df = pd.read_csv("GCN_Cora.csv")
x = df["embedding_dim"]
y = df.iloc[:, 1:]
```

```python
y_max = y.max(axis=1)
y_min = y.min(axis=1)
y_mean = y.mean(axis=1)
```

```python
plt.figure(figsize=(7,5))
plt.plot(x, y_mean)
plt.fill_between(x, y_min, y_max, alpha=0.3)
plt.xlabel("Embedding Dimension")
plt.ylabel("Accuracy")
plt.title("GCN on Cora Dataset")
plt.show()
```

```python
gcn2 = np.load("GCN_Cora_2.npy")
gcn2labels = np.load("GCN_Cora_2_label.npy")
```

```python
plt.figure(figsize=(7,5))
plt.scatter(gcn2[:,0], gcn2[:,1], c=gcn2labels)
plt.title("GCN 2-Dimensional Embeddings")
plt.show()
```

```python
gcn32 = np.load("GCN_Cora_32.npy")
gcn32labels = np.load("GCN_Cora_32_label.npy")
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, perplexity=30, learning_rate=100, n_iter=1000, random_state=0)
gcn32 = tsne.fit_transform(gcn32)
```

```python
plt.figure(figsize=(7,5))
plt.scatter(gcn32[:,0], gcn32[:,1], c=gcn32labels)
plt.title("GCN 32-Dimensional Embeddings")
plt.show()
```
